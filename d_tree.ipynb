```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv(r"D:\sumit_DS project\Datasets with source_code\ML\positon_salaries\Position_Salaries.csv")
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
# Fitting Decision Tree Regression to the datasetmae

from sklearn.tree import DecisionTreeRegressor
regressor1 = DecisionTreeRegressor(criterion="poisson",splitter ='best')
regressor1.fit(X, y)
```

```python
# Predicting a new result
y_pred1 = regressor1.predict([[6.5]])
y_pred1
```

```python
#KNN
from sklearn.neighbors import KNeighborsRegressor
regressor2 = KNeighborsRegressor(n_neighbors=5, weights='uniform', p=2,)     
regressor2.fit(X, y)
```

```python
y_pred2 = regressor2.predict([[6.5]])
y_pred2
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X,regressor1.predict(X), color = 'blue')
plt.title('Truth or bluff (Decision tree Regression)')
plt.xlabel('Position Level')
plt.ylabel('salary')
plt.show()
```

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor1.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor2.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (KNeighborsRegressor)'')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Insight:
1.D_Tree predicts 150k salary for 6.5 years which is no way near the optimum, hence we can say bad model.

2.KNN predicts 168k salary at 6.5 years which is near but still not optimum, hence Average model.
